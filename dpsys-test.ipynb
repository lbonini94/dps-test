```
import pandas as pd
import json
import collections
```

```
js = pd.read_json('https://drive.google.com/u/0/uc?id=1pJIynVU63BK9WGUSKPWTsHDCSPJG5hCb&export=download')
```

```
managers_dict = collections.defaultdict(list)
watchers_dict = collections.defaultdict(list)
```

```
js.sort_values(by='priority', ascending=True)
```

```
for manager in pd.DataFrame(js.managers.to_list())[0].unique():
 for i ,row in js.sort_values(by='priority', ascending=True).iterrows():
    if manager in row.managers:
      managers_dict[str(manager)].append(row['name'])
```

```
dict(managers_dict)
```

```
with open('managers.json', 'w') as fp:
    json.dump(dict(managers_dict), fp)
```

```
a = list(set(list(pd.DataFrame(js.watchers.to_list())[0].unique()) + list(pd.DataFrame(js.watchers.to_list())[1].unique()) + list(pd.DataFrame(js.watchers.to_list())[2].unique()) + list(pd.DataFrame(js.watchers.to_list())[3].unique())))
```

```
for watcher in a:
 for i ,row in js.sort_values(by='priority', ascending=True).iterrows():
    if watcher in row.watchers:
      watchers_dict[str(watcher)].append(row['name'])
```

```
dict(watchers_dict)
```

```
with open('watchers.json', 'w') as fp:
    json.dump(dict(watchers_dict), fp)
```
